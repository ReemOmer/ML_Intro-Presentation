# Decision Tree Model

```python
# Importing required libraries
import matplotlib.pyplot as plt
from sklearn import tree
import numpy as np

# Input
# 0 >> smooth, 1 >> bumby
features = [[140, 1], [130, 1], [150, 0], [170, 0]]

# Output
labels = ['Orange', 'Orange',  'Apple', 'Apple']
```

```python
np.array(features).shape, np.array(labels).shape
```

```python
# ML model, empty rule box
clf = tree.DecisionTreeClassifier()

# Training algorithm (to find patterns in data)
clf = clf.fit(features, labels)
```

```python
# Make prediction based on the learned rules
print (clf.predict([[145,1]]))
```

```python
print(clf.predict([[160,0]]))
```

# Visulization of Decision Trees

```python
fig = plt.figure(figsize=(5,5))
_ = tree.plot_tree(clf, 
                   feature_names=['weight','texture'],  
                   class_names=np.array(['Orange', 'Apple']),
                   filled=True)
```

```python
# Import required libraries
import numpy as np
from sklearn import tree

# Define indexes to delete
test_idx = [0,50,100]

# Training data
train_target = np.delete(iris.target, test_idx)
train_data = np.delete(iris.data, test_idx, axis=0)
```

```python
len(train_data),len(train_target)
```

```python
# Testing data
test_target = iris.target[test_idx]
test_data = iris.data[test_idx]
```

```python
for i in range(3):
    print(test_data[i],test_target[i])
```

```python
# ML model, empty rule box
clf = tree.DecisionTreeClassifier()

# Training algorithm (to find patterns in data)
clf = clf.fit(train_data, train_target)
```

```python
# Print true values
print (test_target)
```

```python
# Test ML model prediction 
print (clf.predict(test_data))
```

```python
iris.target_names
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```
